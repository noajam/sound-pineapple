# Deconvolution
The term deconvolution is often a misnomer applied to transposed convolutions in neural networks. In this context, deconvolutional layer is meant to refer to the reverse of the convolution layer process. In actuality, it is a transposed convolution layer that does much the same as a regular convolutional layer. # Bias Nodes

```python
nn.Conv2d(in_features, out_features, 3, 1, 1, bias=False)
```

A bias can be added to your network layers. It is simply a node that holds the value 1. Treat the bias like the y-intercept value in your regular linear equation.

- $y = mx + b$

Here we have the most basic of neural networks, a perceptron consisting of a single input node and single output node with one weight value applied to the input. A sigmoid activation function is then applied to the result. The output of this network is thus given by the following equation:

- $y = sigmoid(w_{0}x) = \frac{1}{1 + e^{-(w_{0}x)}}$

It is a little easier to read in this format:

- $y = sigmoid(w_{0}x) = \frac{1}{1 + exp(-w_{0}x)}$

![SegmentLocal](images/bI2Tm.gif "segment")

```python
# PyTorch code to represent this network
import torch
import torch.nn as nn

Perceptron = nn.Sequential(
    nn.Linear(1, 1),
    nn.Sigmoid()
    )
```

This is a graph showing the network's output with different weight values.

![SegmentLocal](images/ddyfr.png "segment")

Now, if we add the bias node to the input layer...

![SegmentLocal](images/oapHD.gif "segment")

The complete network function is now this:

- $y = sigmoid(w_{0}x + w_{1})$

Equivalently:

- $y = \frac{1}{1 + e^{-(w_{0}x + w_{1})}}$
- $y = \frac{1}{1 + exp(-(w_{0}x + w_{1}))}$
